```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = [
    "The sky is blue and beautiful",
    "Love this blue and beautiful sky",
    "The quick brown fox jumps over the lazy dog",
    "A king's breakfast has sausages, ham, bacon, eggs, toast and beans",
    "I love green eggs, ham, sausages, and bacon",
    "The brown fox is quick and the blue dog is lazy",
    "The sky is very blue and the sky is very beautiful today"
]

# Initialize TfidfVectorizer with English stop words removed
vectorizer = TfidfVectorizer(stop_words='english')

# Compute the TF-IDF matrix for the documents
tfidf_matrix = vectorizer.fit_transform(documents)

# Get feature names (words)
feature_names = np.array(vectorizer.get_feature_names_out())

# Let's choose the first document for visualization
doc_index = 0
doc_tfidf = tfidf_matrix[doc_index].toarray().flatten()

# Get indices of non-zero TF-IDF scores and sort them by score in descending order
nonzero_indices = np.where(doc_tfidf > 0)[0]
sorted_indices = nonzero_indices[np.argsort(doc_tfidf[nonzero_indices])[::-1]]

# Prepare data for plotting
words = feature_names[sorted_indices]
scores = doc_tfidf[sorted_indices]

# Plot the TF-IDF scores for the chosen document
plt.figure(figsize=(10, 6))
plt.bar(words, scores, color='skyblue')
plt.xlabel("Words")
plt.ylabel("TF-IDF Score")
plt.title("TF-IDF Scores for Document 1")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
